```python
demoalt_df = pd.read_excel("../data/Demographic_Info2_btqedit.xlsx")
#Remove 'PIN' column
#df.drop(['PIN'],axis=1,inplace=True)
demo_df = pd.read_excel("../data/demographics 6-22-16.xlsx")
print demo_df.shape
demo_df.head()
```

### TIPS ID and PIN
Merge demo_df with demoalt_df columns that have the mapping from TIPSID to PIN

```python
#drop excess columns
demoalt_df.drop(demoalt_df.columns[2:],axis=1,inplace=True)
print demoalt_df.shape
demoalt_df.head()
```

```python
print demo_df.columns
print demoalt_df.columns
```

```python
#merge data frames on PIN
demo_df= demo_df.merge(demoalt_df,how='left',left_on=['PIN'],right_on=['PIN'])
print demo_df.shape
demo_df.head()
```

Reformat column names to be lower case and have no spaces or apostrophes

```python
demo_df.rename(columns=lambda x: x.lower().replace(" ","_"),inplace=True)
demo_df.rename(columns={"alzheimer's":'alzheimers'},inplace=True)
demo_df.head()
```

### Hospitalizations and Conditions
clean and binarize/dummify these columns

```python
print demo_df.columns
```

recode the conditions columns from NAN,X to 0,1

```python
for col in [u'chf', u'copd', u'diabetes',u'hypertension', u'stroke',u'hypoglycemia', u'falls',
            u'liver_disease', u'hypotension',u'obesity', u'coronary_artery_disease',u'fractures',
            u'renal',u'alzheimers', u'depression']:
    demo_df[col]= demo_df[col].replace([np.nan, 'X','x'], [0, 1,1])

demo_df.head()    
```

recode the medicaid/hospitalization info from Y,N to 1,0

```python
demo_df['rpm_hospitalized'].value_counts()
```

```python
for col in [u'medicaid',u'hospitalized',u'rpm_hospitalized',
            u'rpm_falls',  u'rehospitalized',u'rpm_rehospitalized']:
    demo_df[col]=demo_df[col].replace(['Y','N','Y '],[1,0,1])
demo_df.head()
```

### Gender, Language, Status, Living Setup
dummify gender, language, status, and living setup. split living setup to Family/Alone CG/no CG

```python
print demo_df['language'].value_counts()
print demo_df['language'].unique()
print demo_df['gender'].value_counts()
```

```python
demo_df['language']=demo_df['language'].replace(['Female'],['Other'])
demo_df['language'].value_counts()
```

```python
#dummify gender and language
demo_df = pd.get_dummies(demo_df,columns=['gender','language'], prefix = ['gender','lang'])
demo_df.head(10)
```

```python
demo_df['status_active']=demo_df['status'].replace(['A','D'],[1,0])
demo_df.head()
```

```python
demo_df['living_set-up'].value_counts()
```

```python
demo_df['living_set-up']=demo_df['living_set-up'].replace(['Alone no CG','Family no CG','Alone with CG'],
                                                          ['Alone No CG','Family No CG','Alone CG'])
demo_df['living_set-up'].value_counts()
```

```python
#demo_df['live_alone']=demo_df['living_set-up'].apply(lambda x: str(x).lower().split()[0])
#demo_df['live_alone']=demo_df['live_alone'].replace(['family','alone','nan'],[0,1,np.nan])
#demo_df['caregiver']=demo_df['living_set-up'].apply(lambda x: str(x).lower().split()[1])
demo_df['caregiver']=demo_df['living_set-up'].replace(['Alone No CG','Family No CG','Alone CG','Family CG'],[0,0,1,1])
demo_df['live_alone']=demo_df['living_set-up'].replace(['Alone No CG','Family No CG','Alone CG','Family CG'],[1,0,1,0])
#def recode_caregiver(val):
#   if np.isnan(val):
#       return np.nan
#    elif str(val).find('No CG'):
#        return False
#    else
#        return True
demo_df.head(10)
```

### Date of Birth
convert to datetime format

```python
demo_df['d.o.b.']=pd.to_datetime(demo_df['d.o.b.'])
demo_df.head()
```
