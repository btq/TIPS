```python
data_df = demo_df.merge(compliance_df,left_on='tips_number',right_on='TIPSID')
data_df.head()
```

```python
data_df['age']=pd.to_datetime(data_df['first_measure_date'])-pd.to_datetime(data_df['d.o.b.'])
data_df.head()
```

```python
new_data = data_df.drop(['pin', 'd.o.b.', 'tips_number', 'posttips_rehospitalized',
                         'missed_wks', 'total_wks', 'first_measure_date', 'status_active','medicaid'], axis = 1)
new_data.dropna(inplace=True)
new_data.reset_index(drop=True,inplace=True)
print new_data.shape
new_data.dtypes
```

```python
print new_data.drop(['posttips_hospitalized'],axis=1).shape
#print new_data['compliance'].shape
new_data['age']=new_data['age'].apply(lambda x: pd.tslib.Timedelta(x).days)
```

```python
continuous_feats = ['age','HeartRate_SP02','SP02','HeartRate','Diastolic','Systolic']
new_data[continuous_feats].hist()
```

```python
new_data[continuous_feats].describe()
```

```python
from IPython.display import display # Allows the use of display() for DataFrames

# For each feature find the data points with extreme high or low values
for feature in continuous_feats:
    
    # TODO: Calculate Q1 (25th percentile of the data) for the given feature
    Q1 = np.percentile(new_data[feature],25)
    
    # TODO: Calculate Q3 (75th percentile of the data) for the given feature
    Q3 = np.percentile(new_data[feature],75)
    
    # TODO: Use the interquartile range to calculate an outlier step (1.5 times the interquartile range)
    step = (Q3-Q1)*1.5
    
    # Display the outliers
    print "Data points considered outliers for the feature '{}':".format(feature)
    display(new_data[~((new_data[feature] >= Q1 - step) & (new_data[feature] <= Q3 + step))])
    
```

```python
# OPTIONAL: Select the indices for data points you wish to remove
age_outliers  = [176,181,183,184,203,205,208,217,223,225,226,228,229,233,241,242,269,276]
spo2_outliers = [90,186,196,247,294,295]
hr_outliers = [306,319,380,381]
outliers = age_outliers + hr_outliers
# Remove the outliers, if any were specified
good_data = new_data.drop(new_data.index[outliers]).reset_index(drop = True)
```

```python
#removing HeartRates less than 10 (4 rows - need to investigate)
new_data = new_data[new_data['HeartRate']>10]
```

```python
from scipy import stats
plt.hist(new_data['SP02'])
xt, _ = stats.boxcox(new_data['SP02'].astype('float'),lmbda=None)
plt.hist(xt)
```

```python
from sklearn.preprocessing import Normalizer
nrmlzr = Normalizer(copy=False)
new_data[continuous_feats] = nrmlzr.fit_transform(new_data[continuous_feats])
```

```python
new_data[continuous_feats].hist()
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(new_data.drop(['posttips_hospitalized'],axis=1), new_data['posttips_hospitalized'], stratify=new_data['posttips_hospitalized'],test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=42)
clf.fit(X_train,y_train)

score = clf.score(X_test,y_test)
print score
```

```python
from sklearn.grid_search import GridSearchCV
#set parameters
logr_parameters = {'penalty':['l2','l1'],'tol':[.001,.0001,.01],
                    'C':[1.0,.9],'max_iter':[100,500]}
#set scoring function (optional)
from sklearn.metrics import f1_score
from sklearn.metrics import make_scorer
f1_scorer = make_scorer(f1_score)

#grid search initialization
grid_obj = GridSearchCV(clf,param_grid = logr_parameters)

# TODO: Fit the grid search object to the training data and find the optimal parameters
grid_obj = grid_obj.fit(X_train,y_train)
```

```python
print grid_obj.best_estimator_
print grid_obj.best_score_
print grid_obj.scorer_
print grid_obj.score(X_test,y_test)
pd.DataFrame(data=grid_obj.best_estimator_.coef_,columns=new_data.drop(['posttips_hospitalized'],axis=1).columns)
```

```python
# TODO: Create a decision tree regressor and fit it to the training set
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=42,max_depth=5)
regressor.fit(X_train,y_train)

score = regressor.score(X_test,y_test)
print score
```
